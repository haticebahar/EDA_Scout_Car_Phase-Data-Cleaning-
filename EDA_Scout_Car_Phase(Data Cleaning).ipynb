Boş olan verileri görmüş olduk. Bu verilerin %90 dan fazlasının olmadığı columnları silelim

```python
def df_nans(df, limit):
    missing=df.isnull().sum()*100/df.shape[0]
    return missing.loc[lambda x: x>= limit]

def column_nans(serial):
    return serial.isnull().sum()*100 / serial.shape[0]
```

```python
df_nans(df,90)

#missing value'ları %90 dan fazla olan  column nameleri döndürdü
```

```python
column_nans(df["Electricity consumption"])
```

```python
drop_columns=df_nans(df,90).index
drop_columns
```

```python
df.drop(drop_columns,axis=1,inplace=True)
```

```python
df.drop("null",axis=1,inplace=True)
```
